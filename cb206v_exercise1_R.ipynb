## power calculator examples

```r
# My function which raises a number to some power
power <- function(my_base, my_exponent) {
  ans <- 1 # A placeholder for the answer
  counter <- 0 # A counter which is initialized at zero

  while (counter < my_exponent) {
    ans <- ans * my_base
    counter <- counter + 1
  }

  return(ans)
}

# Test the function
print(power(2, 8))
```

## EXERCISE 2 - linear algebra



Matrix multiplication is an important operation fundamental to linear algebra that is good to understand for this course. Given two matrices, let's call them $A$ and $B$, where $A$ has dimensions $m \times n$ and $B$ has dimensions $n \times p$, their product, $C = A \times B$, will be a matrix of size $m \times p$.

To compute each element $C_{ij}$ of the resulting matrix $C$, here's what we do:
1. **Take the $i$-th row of matrix $A$**.
2. **Take the $j$-th column of matrix $B$**.
3. **Multiply the corresponding elements** from the row and column.
4. **Sum up** these products to get the element $C_{ij}$.

Let's look at an example to make this clearer:
If we have
\[
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}
\]
and
\[
B = \begin{bmatrix} 5 & 6 \\ 7 & 8 \end{bmatrix}
\]

Then, the product matrix $C$ is:
\[
C = A \times B = \begin{bmatrix} (1 \times 5 + 2 \times 7) & (1 \times 6 + 2 \times 8) \\ (3 \times 5 + 4 \times 7) & (3 \times 6 + 4 \times 8) \end{bmatrix} = \begin{bmatrix} 19 & 22 \\ 43 & 50 \end{bmatrix}
\]

### Why is Matrix Multiplication Important?

Matrix multiplication is used in a lot of machine learning algorithms, for example:

- **Linear Regression**: LAter in this course, we will perform linear regression. This often involves solving equations that can be represented as matrix multiplications.

- **Neural Networks**: In neural networks, each layer of the network applies a linear transformation to the input data, followed by a non-linear activation function. The linear transformation is done using matrix multiplication. The weights of the network, which are learned during training, are organized into matrices, and as the data passes through each layer, it is transformed by multiplying it with these weight matrices.

As we progress, we will be calling high-level functions to perform these operations, but under the hood, these functions are using matrix multiplication to handle complex computations. By understanding the basics of matrix multiplication now, we will be better equipped to understand how these algorithms work at a deeper level.

## Anatomy of a program - Neural network fitting to parabolic data example

```r
library(readr)
library(ggplot2)
library(nnet)

# read data from URL
url <- "https://raw.githubusercontent.com/Intertangler/ML4biotech/main/polynomial_data.txt"
df <- read_delim(url, delim = '\t')
X <- as.matrix(df$X)
y <- as.matrix(df$y)

# define a neural network model
build_and_train_model <- function(X, y, epochs = 15, learning_rate = 0.2, size = 5) {
  X_scaled <- scale(X) # normaliznig
  y_scaled <- scale(y)

  # Train
  model <- nnet(X_scaled, y_scaled, size = size, linout = TRUE, decay = learning_rate, maxit = epochs)

  return(list(model = model, X_scaled = X_scaled, y_scaled = y_scaled))
}

# train
model_info <- build_and_train_model(X, y)
model <- model_info$model
X_scaled <- model_info$X_scaled
y_scaled <- model_info$y_scaled

# predictions
n_test <- 100
predictions <- predict(model, X_scaled[seq(1, length(X_scaled), by = n_test), , drop = FALSE])
predictions <- predictions * attr(y_scaled, "scaled:scale") + attr(y_scaled, "scaled:center")
actual_y <- y[seq(1, length(y), by = n_test)]

# plot on top of each other
plot_df <- data.frame(
  X = c(as.vector(X), as.vector(X[seq(1, length(X), by = n_test)])),
  y = c(as.vector(y), as.vector(predictions)),
  label = c(rep("Actual", length(y)), rep("Prediction", length(predictions)))
)

ggplot(plot_df, aes(x = X, y = y, color = label)) +
  geom_point(aes(size = label)) +
  scale_size_manual(values = c(1, 3)) +
  ggtitle('neural network fit to data') +
  xlab('X') +
  ylab('y') +
  theme_minimal()
```

#### Part 1: Matrix Transpose

The **transpose** of a matrix is obtained by flipping the matrix over its diagonal. This means that the rows of the original matrix become the columns in the transposed matrix, and vice versa.

Mathematically, if $A$ is an $m \times n$ matrix, then its transpose, denoted as $A^T$, is an $n \times m$ matrix such that:

$A^T_{ij} = A_{ji}$

For example, if
$A = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \end{bmatrix}$
then the transpose of $A$ is
$A^T = \begin{bmatrix} 1 & 4 \\ 2 & 5 \\ 3 & 6 \end{bmatrix}$

#### Part 2: Matrix Inverse

The **inverse** of a square matrix $A$, denoted as $A^{-1}$, is the matrix that, when multiplied by $A$, yields the identity matrix $I$. Mathematically, this is represented as:

$A \times A^{-1} = I$

$A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix}$
then its inverse is
$A^{-1} = \frac{1}{(1 \times 4 - 2 \times 3)} \begin{bmatrix} 4 & -2 \\ -3 & 1 \end{bmatrix} = \begin{bmatrix} -2 & 1 \\ 1.5 & -0.5 \end{bmatrix}$



Not all matrices have an inverse. For a matrix to have an inverse, it must be square (having the same number of rows and columns) and its determinant must be non-zero. The determinant is a scalar value that can be computed from the elements of a square matrix.

For a 2x2 matrix:
$ A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$
the determinant is calculated as:
$\text{det}(A) = ad - bc$
