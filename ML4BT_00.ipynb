## power calculator examples

```python
# my function which raises a number to some power
def power(my_base, my_exponent): # line defining the function
  ans = 1 # a placeholder for the answer
  counter = 0 # a counter which is initialized at zero
  while counter < my_exponent:
    ans = ans * my_base
    counter = counter + 1
  return ans

print(power(2,8))
```

## Neural network fitting to parabolic data example

```python
# Importing necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
# Read data from URL
url = "https://raw.githubusercontent.com/Intertangler/ML4biotech/main/polynomial_data.txt"
df = pd.read_csv(url, sep='\t')

# Extract Data for Neural Network
X = df['X'].values.reshape(-1, 1)
y = df['y'].values.reshape(-1, 1)

# Define Neural Network Model
def build_and_train_model(X, y, epochs=50, learning_rate=0.2):
    model = Sequential([
        Dense(5, activation='relu', input_shape=(1,)),
        Dense(5, activation='relu'),
        Dense(1)
    ])

    custom_adam = Adam(learning_rate=learning_rate)
    model.compile(optimizer=custom_adam, loss='mse')
    model.fit(X, y, epochs=epochs)
    return model

# Train the Neural Network
model = build_and_train_model(X, y)

# Make Predictions
predictions = []
n_test = 100
for x in X[0::n_test]:
    prediction = model.predict(np.array([x]))  # Using np.array to match input shape
    predictions.append(prediction[0][0])

# Visualization
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Actual')
plt.scatter(X[0::n_test], predictions, color='red', label='Prediction', s = 300)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Neural Network for Fitting Some Parabolic Data')
plt.legend()
plt.show()
```

```python
%%R
# Load required libraries
library(keras)
library(readr)
library(ggplot2)

# Read data from URL
url <- "https://raw.githubusercontent.com/Intertangler/ML4biotech/main/polynomial_data.txt"
df <- read_delim(url, delim = '\t')
X <- as.matrix(df$X)
y <- as.matrix(df$y)

# Define Neural Network Model and train it
build_and_train_model <- function(X, y, epochs = 50, learning_rate = 0.2) {
  model <- keras_model_sequential() %>%
    layer_dense(units = 5, activation = 'relu', input_shape = c(1)) %>%
    layer_dense(units = 5, activation = 'relu') %>%
    layer_dense(units = 1)

  custom_adam <- optimizer_adam(lr = learning_rate)

  model %>% compile(
    optimizer = custom_adam,
    loss = 'mean_squared_error'
  )

  model %>% fit(X, y, epochs = epochs)

  return(model)
}

# Train the Neural Network
model <- build_and_train_model(X, y)

# Make Predictions
n_test <- 100
predictions <- predict(model, X[seq(1, length(X), by = n_test)])

# Visualization using ggplot2
plot_df <- data.frame(
  X = c(as.vector(X), as.vector(X[seq(1, length(X), by = n_test)])),
  y = c(as.vector(y), as.vector(predictions)),
  label = c(rep("Actual", length(y)), rep("Prediction", length(predictions)))
)

ggplot(plot_df, aes(x = X, y = y, color = label)) +
  geom_point(aes(size = label)) +
  scale_size_manual(values = c(1, 3)) +
  ggtitle('Neural Network for Fitting a Polynomial Function') +
  xlab('X') +
  ylab('y') +
  theme_minimal()
```
